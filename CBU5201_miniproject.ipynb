## Audio Features: 
Mel-Frequency Cepstral Coefficients (MFCCs) are widely used in audio processing as they effectively capture the timbral aspects of audio signals. We will extract 40 MFCC features from each audio recording and compute their mean to represent each sample.
## Text Features: 
We will transcribe the audio recordings to raw text using Whisper and extract textual features using TF-IDF Vectorization.

```python
# Import necessary libraries
import os
import requests
from tqdm import tqdm
import pandas as pd
import librosa
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import whisper
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack


from sklearn.feature_extraction.text import TfidfVectorizer




# Function to extract MFCC features from audio
def extract_audio_features(file_path, duration=300, n_mfcc=40):
    try:
        y, sr = librosa.load(file_path, duration=duration)
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
        mfccs_mean = np.mean(mfccs.T, axis=0)
        return mfccs_mean
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return np.zeros(n_mfcc)

# Function to transcribe audio to text using Whisper
def transcribe_audio_whisper(file_path, model):
    try:
        result = model.transcribe(file_path)
        return result["text"]
    except Exception as e:
        print(f"Error transcribing {file_path}: {e}")
        return ""

# Function to extract textual features using TF-IDF
def extract_text_features(text_series):
    vectorizer = TfidfVectorizer(max_features=500)  # Limiting to top 500 features
    tfidf_matrix = vectorizer.fit_transform(text_series)
    return tfidf_matrix, vectorizer
```

```python
# download functions
def download_file(url, save_path):
    """
    
    """
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()  # 
        with open(save_path, 'wb') as f:
            for chunk in response.iter_content(chunk_size=1024):
                if chunk:
                    f.write(chunk)
        print(f": {save_path}")
    except Exception as e:
        print(f": {url}\n: {e}")

def download_files(url_list, save_dir):
    """
    
    """
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
        print(f": {save_dir}")
    
    for url in tqdm(url_list, desc=""):
        filename = os.path.basename(url)
        save_path = os.path.join(save_dir, filename)
        if not os.path.exists(save_path):
            download_file(url, save_path)
        else:
            print(f": {save_path}")

#  CSV  URL 
csv_url = 'https://raw.githubusercontent.com/CBU5201Datasets/Deception/main/CBU0521DD_stories_attributes.csv'
csv_save_path = './dataset/CBU0521DD_stories_attributes.csv'

#  CSV 
download_file(csv_url, csv_save_path)

#  CSV 
df = pd.read_csv(csv_save_path)
print(df.head())

#  URL
base_audio_url = 'https://raw.githubusercontent.com/CBU5201Datasets/Deception/main/CBU0521DD_stories/'

#  URL 
audio_filenames = df['filename'].tolist()
audio_urls = [base_audio_url + filename for filename in audio_filenames]

# 
download_dir = './dataset/audio_files'

# 
download_files(audio_urls, download_dir)

# 
df['file_path'] = df['filename'].apply(lambda x: os.path.join(download_dir, x))

# 
df['file_exists'] = df['file_path'].apply(os.path.exists)

# 
missing_files = df[df['file_exists'] == False]

if not missing_files.empty:
    print("")
    print(missing_files['file_path'])
    # 
    df = df[df['file_exists'] == True].copy()
    print(f" {len(missing_files)} ")
else:
    print("")
    
    
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split
import pandas as pd
from tqdm import tqdm

# ----------------------------- Configuration -----------------------------

# Paths
download_dir = './dataset/audio_files'  # Directory where all audio files are downloaded
csv_file = './dataset/CBU0521DD_stories_attributes.csv'  # Path to the CSV file with metadata

# Output directories
output_dirs = {
    'train': os.path.join('train', 'audio'),
    'validation': os.path.join('validation', 'audio'),
    'test': os.path.join('test', 'audio')
}

# Split ratios
train_size = 0.7
validation_size = 0.15
test_size = 0.15

# Random state for reproducibility
random_state = 42

# Mapping for labels
label_mapping = {'True Story': 1, 'Deceptive Story': 0}  # Adjust based on your actual labels
# ----------------------------- Helper Functions -----------------------------

def create_directories(directories):
    """
    Create directories if they do not exist.
    """
    for dir_path in directories.values():
        os.makedirs(dir_path, exist_ok=True)
        print(f"Directory created or already exists: {dir_path}")

def copy_audio_files(df_subset, source_dir, target_dir):
    """
    Copy audio files from source to target directory based on the DataFrame subset.
    """
    for _, row in tqdm(df_subset.iterrows(), total=df_subset.shape[0], desc=f"Copying to {target_dir}"):
        src = row['file_path']
        dst = os.path.join(target_dir, row['filename'])
        try:
            shutil.copy2(src, dst)
        except Exception as e:
            print(f"Failed to copy {src} to {dst}. Error: {e}")

def save_subset_csv(df_subset, subset_name):
    """
    Save the DataFrame subset to a CSV file.
    """
    csv_path = f"{subset_name}.csv"
    df_subset.to_csv(csv_path, index=False)
    print(f"Saved {subset_name} data to {csv_path}")


def manage_dataset():
    # Step 1: Load the CSV file
    df = pd.read_csv(csv_file)
    print("CSV file loaded successfully.")
    
    # Step 2: Update the 'file_path' column to absolute paths
    df['file_path'] = df['filename'].apply(lambda x: os.path.join(download_dir, x))
    
    # Step 3: Check for missing audio files
    df['file_exists'] = df['file_path'].apply(os.path.exists)
    missing_files = df[~df['file_exists']]
    if not missing_files.empty:
        print("The following audio files are missing:")
        print(missing_files[['filename', 'file_path']])
        # Optionally, remove missing files from the DataFrame
        df = df[df['file_exists']].copy()
        print(f"Removed {len(missing_files)} samples with missing audio files.")
    else:
        print("All audio files are present.")
    
    # Step 4: Verify 'Story Type' column and map labels
    story_type_col = 'Story_type' 
    if story_type_col is None:
        raise ValueError("DataFrame does not contain 'Story_type' column.")
    
    # Clean 'Story Type' values
    df['Story_Type_clean'] = df[story_type_col]
    # print(f"Unique values in '{story_type_col}' after cleaning: {unique_values}")
    
    # Create mapping dictionary based on unique values
    mapping_dict =  {'True Story': 1, 'Deceptive Story': 0}
    df['y'] = df['Story_Type_clean'].map(mapping_dict)
    
    # Identify any unmapped values
    unmapped = df[df['y'].isnull()]
    if not unmapped.empty:
        print("Unmapped 'Story Type' values found:")
        print(unmapped[story_type_col].unique())
        # Decide how to handle unmapped values. Here, we'll remove them.
        df = df[df['y'].notnull()].copy()
        print(f"Removed {len(unmapped)} samples with unmapped 'Story Type' values.")
    else:
        print("All 'Story Type' values have been successfully mapped.")
    
    # Step 5: Split the DataFrame into train, validation, and test sets
    # First split into train and temp (validation + test)
    train_df, temp_df = train_test_split(
        df,
        test_size=(validation_size + test_size),
        stratify=df['y'],
        random_state=random_state
    )
    # Calculate the proportion of validation and test relative to temp
    relative_val_size = validation_size / (validation_size + test_size)
    validation_df, test_df = train_test_split(
        temp_df,
        test_size=(1 - relative_val_size),
        stratify=temp_df['y'],
        random_state=random_state
    )
    print(f"Data split into {len(train_df)} training, {len(validation_df)} validation, and {len(test_df)} test samples.")
    
    # Step 6: Create output directories
    create_directories(output_dirs)
    
    # Step 7: Copy audio files to respective directories
    copy_audio_files(train_df, download_dir, output_dirs['train'])
    copy_audio_files(validation_df, download_dir, output_dirs['validation'])
    copy_audio_files(test_df, download_dir, output_dirs['test'])
    
    # Step 8: Save CSV files for each subset
    save_subset_csv(train_df, 'train')
    save_subset_csv(validation_df, 'validation')
    save_subset_csv(test_df, 'test')
    
    print("Dataset has been successfully split and organized.")

manage_dataset()

```

```python
# Function to evaluate models using cross-validation
def evaluate_model_cv(model, X, y, skf, model_name="Model"):
    accuracy_scores = []
    precision_scores = []
    recall_scores = []
    f1_scores = []
    
    fold = 1
    for train_index, test_index in skf.split(X, y):
        print(f"Fold {fold}:")
        
        X_train_fold, X_test_fold = X[train_index], X[test_index]
        y_train_fold, y_test_fold = y[train_index], y[test_index]
        
        # Train the model
        model.fit(X_train_fold, y_train_fold)
        
        # Predict
        y_pred = model.predict(X_test_fold)
        
        # Calculate metrics
        acc = accuracy_score(y_test_fold, y_pred)
        precision = precision_score(y_test_fold, y_pred, zero_division=0)
        recall = recall_score(y_test_fold, y_pred, zero_division=0)
        f1 = f1_score(y_test_fold, y_pred, zero_division=0)
        
        print(f"  Accuracy: {acc:.4f}")
        print(f"  Precision: {precision:.4f}")
        print(f"  Recall: {recall:.4f}")
        print(f"  F1-Score: {f1:.4f}\n")
        
        # Append scores
        accuracy_scores.append(acc)
        precision_scores.append(precision)
        recall_scores.append(recall)
        f1_scores.append(f1)
        
        fold += 1
    
    # Average metrics
    print(f"--- {model_name} Average Metrics ---")
    print(f"Accuracy: {np.mean(accuracy_scores):.4f}  {np.std(accuracy_scores):.4f}")
    print(f"Precision: {np.mean(precision_scores):.4f}  {np.std(precision_scores):.4f}")
    print(f"Recall: {np.mean(recall_scores):.4f}  {np.std(recall_scores):.4f}")
    print(f"F1-Score: {np.mean(f1_scores):.4f}  {np.std(f1_scores):.4f}\n")
    
    # Evaluate Random Forest with Cross-Validation
print("Random Forest Evaluation with Cross-Validation:")
evaluate_model_cv(rf_classifier, X_scaled, y, skf, model_name="Random Forest")

# Evaluate SVM with Cross-Validation
print("SVM Evaluation with Cross-Validation:")
evaluate_model_cv(svm_classifier, X_scaled, y, skf, model_name="SVM")

# Evaluate Logistic Regression with Cross-Validation
print("Logistic Regression Evaluation with Cross-Validation:")
evaluate_model_cv(lr_classifier, X_scaled, y, skf, model_name="Logistic Regression")
    
    # Evaluate K-Nearest Neighbors with Cross-Validation
print("K-Nearest Neighbors Evaluation with Cross-Validation:")
evaluate_model_cv(knn_classifier, X_scaled, y, skf, model_name="K-Nearest Neighbors")

# Evaluate Naive Bayes with Cross-Validation
print("Naive Bayes Evaluation with Cross-Validation:")
evaluate_model_cv(nb_classifier, X_scaled, y, skf, model_name="Naive Bayes")

# Evaluate Ensemble Model with Cross-Validation
print("Ensemble Model Evaluation with Cross-Validation:")
evaluate_model_cv(ensemble_model, X_scaled, y, skf, model_name="Ensemble Model")

```

6.2 Hyperparameter Tuning and Cross-Validation
To optimize model performance, we will perform hyperparameter tuning using Grid Search within the cross-validation process for each supervised model.

```python
# Hyperparameter tuning for Random Forest
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

grid_rf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid_rf,
    cv=5,
    n_jobs=-1,
    scoring='f1'
)

print("Performing Grid Search for Random Forest...")
grid_rf.fit(X_scaled, y)
print(f"Best parameters for Random Forest: {grid_rf.best_params_}")
print(f"Best cross-validation F1-score: {grid_rf.best_score_:.4f}\n")

# Hyperparameter tuning for SVM
param_grid_svm = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

grid_svm = GridSearchCV(
    estimator=SVC(probability=True, random_state=42),
    param_grid=param_grid_svm,
    cv=5,
    n_jobs=-1,
    scoring='f1'
)

print("Performing Grid Search for SVM...")
grid_svm.fit(X_scaled, y)
print(f"Best parameters for SVM: {grid_svm.best_params_}")
print(f"Best cross-validation F1-score: {grid_svm.best_score_:.4f}\n")

# Hyperparameter tuning for Logistic Regression
param_grid_lr = {
    'C': [0.1, 1, 10],
    'penalty': ['l2'],
    'solver': ['lbfgs', 'saga']
}

grid_lr = GridSearchCV(
    estimator=LogisticRegression(max_iter=1000, random_state=42),
    param_grid=param_grid_lr,
    cv=5,
    n_jobs=-1,
    scoring='f1'
)

print("Performing Grid Search for Logistic Regression...")
grid_lr.fit(X_scaled, y)
print(f"Best parameters for Logistic Regression: {grid_lr.best_params_}")
print(f"Best cross-validation F1-score: {grid_lr.best_score_:.4f}\n")

# Hyperparameter tuning for K-Nearest Neighbors
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

grid_knn = GridSearchCV(
    estimator=KNeighborsClassifier(),
    param_grid=param_grid_knn,
    cv=5,
    n_jobs=-1,
    scoring='f1'
)

print("Performing Grid Search for K-Nearest Neighbors...")
grid_knn.fit(X_scaled, y)
print(f"Best parameters for KNN: {grid_knn.best_params_}")
print(f"Best cross-validation F1-score: {grid_knn.best_score_:.4f}\n")

# Hyperparameter tuning for Naive Bayes is typically not required as it has fewer parameters
# However, if desired, parameters like var_smoothing can be tuned
param_grid_nb = {
    'var_smoothing': [1e-09, 1e-08, 1e-07]
}

grid_nb = GridSearchCV(
    estimator=GaussianNB(),
    param_grid=param_grid_nb,
    cv=5,
    n_jobs=-1,
    scoring='f1'
)

print("Performing Grid Search for Naive Bayes...")
grid_nb.fit(X_scaled, y)
print(f"Best parameters for Naive Bayes: {grid_nb.best_params_}")
print(f"Best cross-validation F1-score: {grid_nb.best_score_:.4f}\n")

# Update models with best parameters
best_rf = grid_rf.best_estimator_
best_svm = grid_svm.best_estimator_
best_lr = grid_lr.best_estimator_
best_knn = grid_knn.best_estimator_
best_nb = grid_nb.best_estimator_

# Retrain Ensemble with tuned models
ensemble_tuned = VotingClassifier(
    estimators=[
        ('rf', best_rf),
        ('lr', best_lr),
        ('svc', best_svm),
        ('knn', best_knn),
        ('nb', best_nb)
    ],
    voting='soft'
)

# Evaluate Tuned Ensemble Model with Cross-Validation
print("Tuned Ensemble Model Evaluation with Cross-Validation:")
evaluate_model_cv(ensemble_tuned, X_scaled, y, skf, model_name="Tuned Ensemble Model")

```
